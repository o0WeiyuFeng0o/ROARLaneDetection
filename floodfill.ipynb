```python
import numpy as np 
import cv2
import imutils
```

```python
cap = cv2.VideoCapture("/home/roar/Desktop/ROARLaneDetection/training_1000.MOV")
cap.isOpened()
```

```python
while True:
    ret,frame = cap.read()
    if not ret:
        break
    #frame = imutils.resize(frame,width = 960)
    seed = (1079, 960)

    box = (200, 200)
    x = (seed[0] - box[0] - box[1], seed[0])
    y = (seed[1] - box[1], seed[1] + box[1])

    mean = np.mean(frame[x[0]:x[1], y[0]:y[1], :], axis = (0,1))
    std = np.std(frame[x[0]:x[1], y[0]:y[1], :], axis = (0,1))
    scale = 2.5
    scale_std = std * scale
    
    print(mean, '/n',std)

    frame[seed[0], seed[1]] = mean

    cv2.floodFill(frame, None, seedPoint=seed, newVal=(0, 0, 255), loDiff=tuple(scale_std), upDiff=tuple(scale_std), flags= cv2.FLOODFILL_FIXED_RANGE)
    #cv2.circle(frame, seed, 2, (0, 255, 0), cv2.FILLED, cv2.LINE_AA);

    cv2.imshow('image',frame)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        cv2.destroyAllWindows()
        break
```
